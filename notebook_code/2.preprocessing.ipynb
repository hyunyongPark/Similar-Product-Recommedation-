```python
# str to dict
df['text'] = [literal_eval(df['text'][i]) for i in tqdm.tqdm(range(len(df)))]
df
```

```python
df['txt_keys'] = [list(df['text'][i].keys()) for i in tqdm.tqdm(range(len(df)))]
df['txt_values'] = [list(df['text'][i].values()) for i in tqdm.tqdm(range(len(df)))]
```

```python
df['length'] = [len(df.txt_keys[i]) for i in tqdm.tqdm(range(len(df)))]
df
```

```python
fit_ls = []
for idx,i in enumerate(df.text):
    try:
        if i['핏'] is not None:
            fit_ls.append((idx,i['핏']))
    except:
        pass

print_ls = []
for idx,i in enumerate(df.text):
    try:
        if i['프린트'] is not None:
            print_ls.append((idx,i['프린트']))
    except:
        pass

color_ls = []
for idx,i in enumerate(df.text):
    try:
        if i['색상'] is not None:
            color_ls.append((idx,i['색상']))
    except:
        pass
```

```python
color_ids = [x[0] for x in color_ls]
print_ids = [x[0] for x in print_ls]
fit_ids = [x[0] for x in fit_ls]
ids = list(set(color_ids)&set(print_ids)&set(fit_ids)) #final idx
df_t = df.iloc[ids,:]
```

```python
fit_ls = []
print_ls = []
color_ls = []
for i in df_t['text']:
    if i['프린트'] is not None:
        print_ls.append(i['프린트'])
    if i['색상'] is not None:
        color_ls.append(i['색상'])
    if i['핏'] is not None:
        fit_ls.append(i['핏'])
    else:
        print(i)

df_t['color'] = color_ls
df_t['fit'] = fit_ls
df_t['print'] = print_ls
```
